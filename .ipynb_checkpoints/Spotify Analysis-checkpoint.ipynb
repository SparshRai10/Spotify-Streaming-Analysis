### Phase 1 - Importing Libraries

```python
#Starting with importing the necessary libraries
import pandas as pd
```

### Phase 2 - Loading the dataset

```python
df = pd.read_csv("Most_Streamed_Spotify_Songs_2024.csv", encoding='latin1')
```

### Phase 3 - Data Preprocessing

```python
df.info() #To get info about our data
```

```python
df.isnull().sum()   #To check for total number of missing/Null values in each column.
```

```python
#Dropping all the columns which have high amount of missing values
df.drop(['TIDAL Popularity','Soundcloud Streams','TikTok Posts','TikTok Likes', 'TikTok Views', 'YouTube Playlist Reach','SiriusXM Spins'
        ,'Deezer Playlist Count','Deezer Playlist Reach','Amazon Playlist Count','Pandora Streams','Pandora Track Stations'], axis = 1, inplace = True)
```

```python
#Checking for duplicate rows
duplicate_rows = pd.DataFrame(df[df.duplicated()].index)
print(duplicate_rows)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
#Removing commas ',' and spaces within the values
df = df.map(lambda x: x.replace(',', '') if isinstance(x, str) else x)
df
```

```python
#Dropping 5 rows which have missing 'Artist' 
df.dropna(subset=['Artist'], inplace = True)
```

```python
#Converting the 'Release Date' into datetime datatype
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')
```

```python
# Extracting Year and Month from the 'Release Date'
df['Year'] = df['Release Date'].dt.year
df['Month'] = df['Release Date'].dt.month
```

```python
df['Explicit Track'] = df['Explicit Track'].astype(bool)
df['Explicit Track'] = df['Explicit Track'] == 1
```

```python
# Converting Integer value columns to numeric (integer type) as they are currently stored as objects.
columns_to_convert = ['All Time Rank', 'Spotify Streams', 'Spotify Playlist Count', 
                      'Spotify Playlist Reach', 'YouTube Views', 'YouTube Likes', 
                      'AirPlay Spins', 'Shazam Counts']

for col in columns_to_convert:
    # Converting to numeric, coerce the errors to NaN
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Replace NaN values with the mean of each column and convert to int (we are replacing missing values with the 'mean' values).
for col in columns_to_convert:
    df[col] = df[col].fillna(df[col].mean()).astype(int)

# For float columns, replace NaN with the mean and convert to int (as decimals are not needed)
df['Track Score'] = df['Track Score'].fillna(df['Track Score'].mean()).astype(int)
df['Spotify Popularity'] = df['Spotify Popularity'].fillna(df['Spotify Popularity'].mean()).astype(int)
df['Apple Music Playlist Count'] = df['Apple Music Playlist Count'].fillna(df['Apple Music Playlist Count'].mean()).astype(int)
print(df.dtypes)
```

### Phase 4 - EDA and Data Visualization

##### Top 20 Streaming Songs on Spotify (all time)

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Grouping by songs and taking the sum of Spotify Streams
top_20_spotify_songs = df.groupby('Track', as_index=False)['Spotify Streams'].sum()

# Sorting by streams in descending order and taking the top 20 songs
top_20_spotify_songs = top_20_spotify_songs.sort_values('Spotify Streams', ascending=False).head(20)

# Displaying Spotify Streams in Billions
top_20_spotify_songs['Spotify Streams (in Billion)'] = top_20_spotify_songs['Spotify Streams'] / 1000000000

# Plotting the barplot
plt.figure(figsize=(10, 6))
ax = sns.barplot(y='Track', x='Spotify Streams (in Billion)', data=top_20_spotify_songs, hue='Track', palette='magma', dodge=False, legend=False)
plt.title('Top 20 Streaming songs on Spotify')
plt.xlabel('Total Spotify Streams (in Billions)')
plt.ylabel('song')
plt.show()
```

##### Top 20 Streaming Artists on Spotify (of all time)

```python
top_20_spotify_artists = df.groupby('Artist', as_index=False)['Spotify Streams'].sum()
top_20_spotify_artists = top_20_spotify_artists.sort_values('Spotify Streams', ascending=False).head(20)
top_20_spotify_artists['Spotify Streams (in Billion)'] = top_20_spotify_artists['Spotify Streams'] / 1000000000
# Plotting
plt.figure(figsize=(10, 6))
ax = sns.barplot(y='Artist', x='Spotify Streams (in Billion)', data=top_20_spotify_artists, hue='Artist', palette='magma', dodge=False, legend=False)
plt.title('Top 20 Streaming Artists on Spotify (of all time)')
plt.xlabel('Total Spotify Streams (in Billions)')
plt.ylabel('Artist')
plt.show()
```

##### Number of Songs by each top Artist

```python
#Taking the list of top 20 artists from the above cell output
list_of_top_20_artists = [
    'Bad Bunny', 'Taylor Swift', 'Drake', 'MUSIC LAB JPN', 'Ariana Grande',
    'The Weeknd', 'Doja Cat', 'Bruno Mars', 'Maroon 5', 'Eminem', 'Rihanna',
    'KAROL G', 'Miley Cyrus', 'Lil Nas X', 'XXXTENTACION', 'Billie Eilish',
    'Justin Bieber', 'Travis Scott', 'Coldplay', 'Dua Lipa'
]

filtered_set = df[df['Artist'].isin(list_of_top_20_artists)]
Top_Artist_songs_count = filtered_set['Artist'].value_counts().reset_index()
Top_Artist_songs_count.columns = ['Artist', 'Count']


plt.figure(figsize=(10, 6))
colors = sns.color_palette("magma", len(Top_Artist_songs_count))
# Creating the horizontal bar plot
bars = plt.barh(Top_Artist_songs_count['Artist'], Top_Artist_songs_count['Count'], color=colors)
for bar in bars:
    plt.text(
        bar.get_width() + 1,  # Positioning text to the right of the bar
        bar.get_y() + bar.get_height() / 2,  # Centering the text vertically
        int(bar.get_width()),  # The count value
        va='center',  # Vertical alignment
        fontsize=9  # Font size
    )
plt.xlabel('Song Count')
plt.ylabel('Artist')
plt.title('Number of Songs by each top Artist')
plt.gca().invert_yaxis()  # Invert y-axis to have the artist with the highest count on top
plt.show()
```

##### Top 10 most listened Artists on Spotify in the year 2024

```python
year2024 = df.copy()
year2024 = year2024.loc[df['Year'] > 2023] 
year2024.drop_duplicates(subset='Track', keep='first', inplace=True)
```

```python
top_10_spotify_artists_2024 = year2024.groupby('Artist', as_index=False)['Spotify Streams'].sum()
top_10_spotify_artists_2024 = top_10_spotify_artists_2024.sort_values('Spotify Streams', ascending=False).head(20)
# Convert Spotify Streams to Billions
top_10_spotify_artists_2024['Spotify Streams (in Billion)'] = top_10_spotify_artists_2024['Spotify Streams'] / 1000000000
# Plotting
plt.figure(figsize=(10, 6))
ax = sns.barplot(y='Artist', x='Spotify Streams (in Billion)', data=top_10_spotify_artists_2024, hue='Artist', palette='magma', dodge=False, legend=False)
plt.title('Top 10 most listened Artists on Spotify in the year 2024')
plt.xlabel('Total Spotify Streams (in Billions)')
plt.ylabel('Artist')
plt.show()
```

##### Spotify Popularity Vs Track Score (for each top 10 Track of 2024)

```python
top_10_spotify = year2024.sort_values(by='Spotify Streams',ascending=False).head(10)
df_melted = pd.melt(top_10_spotify,id_vars='Track',value_vars=['Track Score','Spotify Popularity'],var_name='Score Popularity',value_name='count')
plt.figure(figsize=(16, 10))
sns.barplot(x='Track', y='count', hue='Score Popularity', palette='magma',data=df_melted)
plt.title('Spotify Popularity Vs Track Score (for each top 10 Track of 2024)',fontsize=20)
plt.xlabel('Track',fontsize=16)
plt.ylabel('count',fontsize=16)
plt.xticks(rotation=90,fontsize=16)
plt.yticks(fontsize=16)
plt.tight_layout()
plt.show()
```

##### Number of tracks released every year till 2024

```python
tracks_per_year = df.groupby('Year')['Track'].count().reset_index(name='tracks_per_year')
plt.figure(figsize=(14, 6))
sns.lineplot(data=tracks_per_year, x='Year', y='tracks_per_year', marker='o', color='magenta')
plt.title('Number of Tracks Every Year')
plt.xlabel('Year')
plt.ylabel('Number of Tracks')
plt.show()
```

###### Conclusion:
Number of tracks released are increasing every year, a huge increase is noticed between the years 2015-2024

##### Spotify streaming curve (till year 2024)

```python
streams_per_year = df.groupby('Year')['Spotify Streams'].sum().reset_index(name='streams_per_year')
streams_per_year['streams_per_year'] = streams_per_year['streams_per_year'] / 1e9
plt.figure(figsize=(14, 6))
sns.lineplot(data=streams_per_year, x='Year', y='streams_per_year', marker='o', color='magenta')
plt.title('Number of Streams Every Year')
plt.xlabel('Year')
plt.ylabel('Number of streams (in Billions)')
plt.show()
```

###### Conclusion:
Spotify streaming has experienced a consistent increase since 2014; however, it has also experienced a decline after 2023. This decline may be attributed to a variety of factors, including market competition and changes in user preferences.


##### Relation between Track score and Explicitness

```python
explicit_track_scores = df.groupby('Explicit Track')['Track Score'].mean().reset_index(name='explicit_track_scores')
plt.figure(figsize=(10, 5))
sns.barplot(x='Explicit Track',y='explicit_track_scores',data=explicit_track_scores,hue='Explicit Track',dodge=False,legend = False)
plt.title('Track Score depending on Explicitness')
plt.xlabel('Explicit')
plt.ylabel('Average Track Score')
plt.show()
```

###### Conclusion:
Explicit songs are found to have a higher track score than non-explicit music. This could be due to the factor that youth
likes to watch and listen to explicit content.

##### Number of Explicit Tracks per Year

```python
explicit_tracks = df[df['Explicit Track'] == 1]
explicit_tracks_per_year = explicit_tracks.groupby('Year')['Track'].count().reset_index()
explicit_tracks_per_year.columns = ['Year', 'Count']
plt.figure(figsize=(14,5))
sns.lineplot(data=explicit_tracks_per_year,x='Year',y='Count',marker='o')
plt.title('Number of Explicit Tracks per Year')
plt.xlabel('Year')
plt.ylabel('Number of Tracks')
plt.grid(True)
plt.show()
```

##### Explicitness and Track score of top 20 tracks

```python
Track_Score = df.sort_values(by='Track Score', ascending=False).head(20)
plt.figure(figsize=(12, 12))
sns.barplot(x='Track Score', y='Track', data=Track_Score, hue='Explicit Track', dodge=False)
plt.xlabel('Track Score')
plt.ylabel('Track')
plt.title('Explicitness and Track score of top 20 tracks', fontsize=20)
plt.legend(title='Explicit Track', loc='upper right', bbox_to_anchor=(1.12,1))
for index, value in enumerate(Track_Score['Track Score']):
    plt.text(value, index, f'{value}', va='center',ha='right', fontsize=12, color='white')
plt.show()
```

##### Explicit Track distribution per year from 2020 till 2024

```python
plt.figure(figsize=(12,6))
year_distribution = df.loc[df['Year'] > 2019]
sns.boxplot(x='Explicit Track',y='Year',data=year_distribution, palette = 'magma', hue = 'Explicit Track', legend =False)
ax.set_title('Explicit Track distribution per year from 2020',fontsize=20)
plt.yticks(year_distribution['Year'],fontstyle='normal')
plt.show()
```

The distribution of explicit tracks shows a distinct rising pattern between 2020 and 2024.
